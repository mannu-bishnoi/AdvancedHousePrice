# Missing Values

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4), '% missing values')

```

#### Since there are many missing values we need to find the relationship between missing values and sales price

```python
for feature in features_with_na:
    data=dataset.copy()
    
    #let's make a vaiable that indicates 1 if he observation was missing or 0 if present
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    #lets calculate mean SalesPrice where the info is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing values and dependent variable is clearly visible So we need to replace these nan values with some meaningfull data

### Numerical Varaibles

```python
#list of numerical variables
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype != 'O']  # O means Object

print('Number of numerical Variables: ',len(numerical_features))

dataset[numerical_features].head()
```

### Temporal Variables(Eg. DateTime Variable)

From the dataset we have 4 year variable. We have Extract info from the datetime varibale like no. of years or no. of days. One example in this specific scenerio can be difference in years between the year of house was built and year the house was sold.

```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# Let's explore content in these year-feature
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
## let's analyze the temperal DateTime Variable
## we will check wheather there is a relation between year the house is sold and Sales price

data.groupby('YrSold')['SalePrice'].median().plot()
plt.title('House Price vs YearSold')
plt.xlabel('YearSold')
plt.ylabel('House Price Median')
plt.show()
```

Here the price is deceasing as the year increases.

```python
## Compare the diff b/w All years features with sales Price

for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        ## capture the diff b/w yrsold and others
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

### Continous Variable

```python
## Analyze the continous values by histogram to understand the distribution

for feature in continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

### Using logarithmic transformation

```python
for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():    ## pass this beacuse log 0 tends to infinity
        pass
    else:
        data[feature]=np.log(data[feature])  ## log normal transformation
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
```

#### Missing Values
Captutre all nan values
handle categorical features which are missing

```python
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=='O']

for feature in feature_nan:
    print("{}:{}%".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
def replace_nan_cat(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('Missing')
    return data

dataset=replace_nan_cat(dataset,feature_nan)

dataset[feature_nan].isnull().sum()
```

```python
## Check for numerical missing values
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
for feature in numerical_with_nan:
    print('{}:{}%'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replace nan values with median
for feature in numerical_with_nan:
    median_value=dataset[feature].median()
    #new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)  ## for further use weather there was a missing value in which row 1 or 0 for no missing value
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.columns
```

### Temperal Variables

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
    
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

### Numerical Values
are skewed we will perform normal distribution 

```python
continous_feature
```

```python
num_feature=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']  ##only for these beacuse log 0 will be infinity for other features

for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset[num_feature].head()
```

#### Handling Rare Categorical Feature
remove categorical variables that are less than 1%

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head()
```

```python
for feature in categorical_features:
    print('{} : {}'.format(feature,len(dataset[feature].unique())))
```

```python
for feature in categorical_features:
    label_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    label_ordered={k:i for i,k in enumerate(label_ordered,0)}
    dataset[feature]=dataset[feature].map(label_ordered)
```

### Feature Scaling

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
```

```python
len(scaling_feature)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[scaling_feature])
```
